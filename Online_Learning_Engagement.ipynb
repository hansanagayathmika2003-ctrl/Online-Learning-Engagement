Finding Outliers

```python
numeric_cols = ['age', 'internet_speed_mbps', 'study_hours_weekly', 'login_frequency_weekly', 'avg_session_duration_min', 'video_watch_time_min', 'assignments_submitted','forum_posts', 'quiz_attempts', 'avg_quiz_score', 'attendance_rate', 'engagement_score', 'final_grade', 'dropout' ]
```

```python
Q1 = df[numeric_cols].quantile(0.25)
Q3 = df[numeric_cols].quantile(0.75)
IQR = Q3 - Q1
```

```python
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
```

```python
outliers = ((df[numeric_cols] < lower_bound) | (df[numeric_cols] > upper_bound))
```

```python
df_outliers = df[outliers.any(axis=1)]
print(df_outliers)
```

```python
df.shape
```

```python
print(len(df_outliers))
```

```python
print(f"{len(df_outliers)/len(df)*100:.2f}%")
```

```python
for col in numeric_cols:
    col_outliers = df[(df[col] < lower_bound[col]) |
                      (df[col] > upper_bound[col])]

    if len(col_outliers) > 0:
        print(f"\nOutliers in {col}:")
        print(col_outliers[[col]])
```

```python
df.shape
```

```python
print(df['final_grade'][83])
```

```python
df.columns
```

```python
outlier_percent = len(df_outliers) / len(df) * 100
print(f"Outlier %: {outlier_percent:.2f}%")
```

```python
df[numeric_cols] = df[numeric_cols].clip(
                   lower_bound, upper_bound, axis=1)

print("Outliers capped! ✅")
```
